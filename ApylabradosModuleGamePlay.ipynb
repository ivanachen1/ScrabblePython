```python
%cd /content/drive/MyDrive/Colab Notebooks/Python de la A a la Z/Proyecto Final/Depurado
import ApylabradosModuleLogic as ap

class Partida():

  valor = 0

  def dar_bienvenida(self):
    """
    Mensaje de bienvenida
    """
    archivo = "/content/drive/MyDrive/Colab Notebooks/Python de la A a la Z/Proyecto Final/Depurado/mensaje_bienvenida.txt"
    with open(archivo,mode = "r") as lector:
      print(lector.read())

  
  def dar_instrucciones(self):
    """
    Mensaje de bienvenida
    """
    archivo = "/content/drive/MyDrive/Colab Notebooks/Python de la A a la Z/Proyecto Final/Depurado/instrucciones.txt"
    with open(archivo,mode = "r") as lector:
      print(lector.read())

  def repartir_siete_fichas_jugador(self):
    for i in range(7):
      ficha_quitada = bolsa_de_fichas.tomar_ficha_aleatoria()
      fichas_del_jugador.agregar_ficha(ficha_quitada)
  
  @staticmethod
  def ofrecer_ayuda_palabra(objeto_tablero,fichas_del_jugador,lista_palabras_usadas = ""):
    
    print("Len del tablero",len(objeto_tablero.tablero))
    
    if objeto_tablero.total_fichas == 0:

      ap.Diccionario.mostrar_posibles_palabras(fichas_del_jugador)
    
    else:

      # Tengo que construir la nueva palabra con las fichas que tengo mas las que tengo en el tablero

      for i in range(15):
        for j in range(15):
          if objeto_tablero.tablero[i,j] == " ":
            continue
          else:
            caracter = objeto_tablero.tablero[i,j]
            ap.Diccionario.mostrar_posibles_palabras_plus(fichas_del_jugador,caracter,lista_palabras_usadas)

  def ofrecer_ayuda_posicion(self):

      palabra = ap.Palabra.leer_palabra()
      
      if tablero.total_fichas == 0:
        posibilidades = ap.Diccionario.mostrar_posibles_palabras(fichas_del_jugador)

      else:
        posibilidades = ap.Diccionario.mostrar_posibles_palabras_plus(fichas_del_jugador,caracter,lista_palabras_usadas)

      tablero.mostrar_posibles_colocaciones_palabra(fichas_del_jugador,palabra,posibilidades,lista_palabras_usadas)
      

  @staticmethod
  def reponer_fichas_al_jugador(fichas_del_jugador,bolsa_de_fichas):

    while fichas_del_jugador.obtener_total_fichas() < 7:
      fichas_del_jugador.agregar_ficha(bolsa_de_fichas.tomar_ficha_aleatoria())

    return fichas_del_jugador.mostrar_fichas()
  
  
  def proceso_colocar_fichas(self,objeto_fichas_jugador,objeto_tablero):
    """
    Pasos del proceso:
      1) crear un objeto palabra y transformarla a objeto frecuencia
      2) obtener las fichas del jugador y transformarlas a objeto frecuencia
      3) Ver si la palabra se puede formar con las fichas que tengo
      4) Valido si la palabra puede colocarse en el tablero
      5) coloco las fichas en el tablero
      6) descuento las fichas del jugador
      7) aumento los puntos del jugador

    """

    # Cuando no tengo fichas en el tablero
    
    global nueva_palabra
    nueva_palabra = ap.Palabra.leer_palabra()
    proceso_completo = False

    while proceso_completo == False:
      #Con este While permito que el jugador repita la secuencia del juego

      if ap.Diccionario.verificar_existencia_palabra_diccionario(nueva_palabra) != True:
        print("La palabra no está en el diccionario")    
      
      else:

        x = int(input("Por favor introduzca una coordenada x del 1 al 15: "))
        y = int(input("Por favor introduzca la coordenada y del 1 al 15: "))
        orientacion = input("Por favor coloque la orientacion siendo H o V: ")
        orientacion.upper()

        verificacion = input("¿Estas seguro?")
        verificacion.upper()

        while verificacion == "NO":
          x = int(input("Por favor introduzca una coordenada x del 1 al 15: "))
          y = int(input("Por favor introduzca la coordenada y del 1 al 15: "))
          orientacion = input("Por favor coloque la orientacion siendo H o V: ")
          orientacion.upper()
          verificacion = input("¿Estas seguro?")
          verificacion.upper()
        
        frecuencia_palabra = nueva_palabra.get_frecuencia_palabra()
        frecuencia_fichas_del_jugador = objeto_fichas_jugador.get_frecuencia()
        
        if objeto_tablero.total_fichas == 0:
          
          if objeto_tablero.es_posible(objeto_palabra = nueva_palabra, coordenada_x = x -1 , coordenada_y = y-1,direccion = orientacion)[0] == True:
            print("Resultado Es posible",objeto_tablero.es_posible(objeto_palabra = nueva_palabra, coordenada_x = x -1 , coordenada_y = y-1,direccion = orientacion)[0])
            
            
            if ap.Tabla_de_Frecuencias.es_subconjunto_de(frecuencia_palabra,frecuencia_fichas_del_jugador) == True:
              palabra_usada = objeto_tablero.colocar_palabra(objeto_fichas_jugador,palabra = nueva_palabra,coordenada_x = x -1,coordenada_y = y - 1, direccion = orientacion)
              lista_palabras_usadas.append(palabra_usada)
              proceso_completo = True
            else:
              print("No dispones de las fichas para formar la palabra")
          
          else:
            print("No es posible colocar la palabra donde indicaste")  

        elif objeto_tablero.total_fichas >= 1:
          #NECESIDAD: Necesito que las fichas del tablero se incorporen a las fichas del jugador 
          #para poder verificar que pueda crear la palabra
          lista_palabras = ap.Diccionario.lista_completa_palabras(objeto_fichas_jugador,objeto_tablero)
          if nueva_palabra.convert_palabra_to_string() not in lista_palabras:
            print("No dispones de las fichas suficientes para formar la palabra,vos o el tablero no tienen las letras necesarias")
          else:
            #verifico si es posible colocar la palabra, ya se que tiene las fichas necesarias
            print("Primer nivel")
            print("Coordenada X",x-1)
            print("Coordenada Y",y-1)
            print("Resultado Es posible",objeto_tablero.es_posible(objeto_palabra = nueva_palabra, coordenada_x = x -1 , coordenada_y = y-1,direccion = orientacion)[0])
            if objeto_tablero.es_posible(objeto_palabra = nueva_palabra, coordenada_x = x -1 , coordenada_y = y-1,direccion = orientacion)[0] == False:
              print("No es posible colocar las fichas en la posicion indicada")
            else:
              palabra_usada = objeto_tablero.colocar_palabra(objeto_fichas_jugador,palabra = nueva_palabra,coordenada_x = x -1,coordenada_y = y - 1, direccion = orientacion)
              lista_palabras_usadas.append(palabra_usada)
              proceso_completo = True

  
  @classmethod
  def menu_opciones(cls):
    
    print("-------------------------------------")
    print("")
    print("Elija las siguientes opciones")
    print("Consultar tus puntos = 1")
    print("Consultar el valor de una ficha = 2")
    print("Consultar el valor de todas las fichas = 3")
    print("Ver el tablero = 4")
    print("Colocar una palabra = 5")
    print("Pedir sugerencias de palabra = 6")
    print("Pedir ayuda de donde colocar las fichas = 7")
    print("Salir del juego = 8")
    print("")
    print("")

    
  @classmethod
  def definir_valor(cls):

    """
    Le pido la opcion al usuario
    """
   
    valor = input("¿Que opcion desea realizar?")

    while valor not in ["1","2","3","4","5","6","7","8"]:
      print("El valor introducido es incorrecto")
      valor = input("¿Que opcion desea realizar?")
    
    valor = int(valor)
    
    
    while type(valor) != type(1) or valor < 1 or valor > 8:

      print("dato mal tipiado")
      valor = int(input())   
  
    return valor 
  
  @classmethod
  def iniciar_variables(cls):

    global tablero
    tablero = ap.Tablero()   
    ap.Tablero.puntuacion = 0
    
    global fichas_del_jugador
    fichas_del_jugador = ap.Fichas()
    
    global bolsa_de_fichas
    bolsa_de_fichas = ap.Fichas()

    global partida
    partida = Partida()
  
    global lista_palabras_usadas
    lista_palabras_usadas = []

  @classmethod
  def jugar(cls):

    
    opcion = input("Desea iniciar el juego(coloque SI o NO): ")
    play = False
    
    Partida.iniciar_variables()

    while type(opcion) != type("s"):
      print("Ha introducido una palabra invalida,acordate de poner si o no")
      opcion = input("Desea iniciar el juego(coloque SI o NO): ")

    opcion = opcion.lower()

    while opcion != "si" and opcion != "no":
      print("Ha introducido una palabra diferente de si o no, acordate de poner SI o No")
      opcion = input("Desea iniciar el juego(coloque SI o NO): ")

    if opcion == "no":
      print("Ha salido del juego")
    
    elif opcion == "si":
      
      play = True
      partida.dar_bienvenida()
      partida.dar_instrucciones()
      Partida.menu_opciones()
      
      bolsa_de_fichas.crear_fichas()
      
      if len(fichas_del_jugador.letras) < 7:
        print("TUS FICHAS SON")
        Partida.reponer_fichas_al_jugador(fichas_del_jugador,bolsa_de_fichas)

      if len(fichas_del_jugador.letras) < 7:
        print("TUS FICHAS SON")
        Partida.reponer_fichas_al_jugador(fichas_del_jugador,bolsa_de_fichas)
           
      valor = Partida.definir_valor()

      while play == True and valor != 8:

        if len(fichas_del_jugador.letras) < 7:
          print("TUS FICHAS SON")
          Partida.reponer_fichas_al_jugador(fichas_del_jugador,bolsa_de_fichas)
          
          fichas_del_jugador.mostrar_fichas()
          print("")
        
        #else:
        #  print("TUS FICHAS SON")
        #  fichas_del_jugador.mostrar_fichas()
      
        
        #Anda  
        if valor == 1:
          print("Tus puntos son: {}".format(ap.Tablero.puntuacion))
          print("")         
          valor = Partida.definir_valor()
      
        #Anda
        elif valor == 2:
          ficha = input("Introduzca la ficha que quiera consultar: ")
            
          while type(ficha) != type("s"):
            print("coloque una letra valida")
            ficha = input("Introduzca la ficha que quiera consultar: ")

          while len(ficha) != 1:
            print("Ha colocado mas de una palabra")
            ficha = input("Introduzca la ficha que quiera consultar: ")

          ficha = ficha.upper()
          print("")
          if ap.Fichas.obtener_puntos(ficha) == 1:
            print("La letra {} vale {} punto".format(ficha,ap.Fichas.obtener_puntos(ficha)))
          elif ap.Fichas.obtener_puntos(ficha) > 1:
            print("La letra {} vale {} puntos".format(ficha,ap.Fichas.obtener_puntos(ficha)))
          else:
            print("La letra no esta en el diccionario")
          print("")
        
          valor = Partida.definir_valor()

        #Anda
        elif valor == 3:
          ap.Fichas.mostrar_puntos_fichas()      
          valor = Partida.definir_valor()

        #Anda
        elif valor == 4:
          tablero.mostrar_tablero()
          print("")
          valor = Partida.definir_valor()


        elif valor == 5:
          partida.proceso_colocar_fichas(objeto_tablero = tablero,objeto_fichas_jugador = fichas_del_jugador)
          print("")
          valor = Partida.definir_valor()                
        #Anda
        elif valor == 6:
          
          resultado = Partida.ofrecer_ayuda_palabra(tablero,fichas_del_jugador,lista_palabras_usadas)
          valor = Partida.definir_valor()        
        
        elif valor == 7:
          partida.ofrecer_ayuda_posicion()
          print("")         
          valor = Partida.definir_valor()
        #Anda
        elif valor == 8:
          print("Has salido del juego")
          print("Tu puntuacion final ha sido de {} puntos".format(ap.Tablero.puntuacion))
          play = False
           
```
